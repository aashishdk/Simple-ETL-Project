### Buiding ETL pipeline
step1 : Extract data from CSV file into pandas dataframe
step2 : Transform data (i.e cleaning the data)
step3 : Create Database
step4 : Load clean data into database


```python
import pandas as pd  # for data ETL
import psycopg2 # for connecting python file and postgreesssql
from sqlalchemy import create_engine #to efficiently manage and resuse database connections
```

### Extract data from csv file


```python
df = pd.read_csv(r"F:\DE Project\dataset\Softwork Technologies - Sheet1.csv")
df.head()
```

```python
# Transform data(i.e Cleaning data)
df.duplicated().sum()
```

```python
# Dropping duplicates value from the csv files
df.drop_duplicates(keep='first',inplace=True)
```

```python
# Checking whether columns contains null values or not if it,it returns true
df.isnull()
```

```python
#Again Checking whether duplicated value is removed or not
df.duplicated().sum()
```

```python
# Checking total number of null values present in the each column field
df.isnull().sum()
```

```python
### Dealing with missing values in education and previous_year_rating value
df['education'] = df['education'].fillna('unknown')
```

```python
df['previous_year_rating'] = df['previous_year_rating'].fillna(0)
```

```python
df.isnull().sum()
```
